## args/kwargs

```python
#dynamic parameter sender

def func1(x,y):
    return x + y

print(func1(10,20))
print(func1(1100, 200))

```

```python
def func1(x,y,z = 10000):
    return x + y

print(func1(10,20))
print(func1(1100, 200, 500))
```

```python
#Dont know how many parameters in funciton def or function call

#args

def func1(*args):
    print(args)
    sum1 = 0
    for i in args:
        sum1 = sum1 + i
    return sum1

print(func1(10,20))
print(func1(1100, 200, 500))
print(func1(1100, 200, 500,5000,6,3,4,5))
```

```python
#kwargs ==> dynamic dictionary or key value pair

def func1(**kwargs):
    print(kwargs)
    

print(func1(var1=10,var2=20))
print(func1(var3=1100, var4=200, var5=500))
```
